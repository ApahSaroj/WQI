```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/DATA/Taladanda canal data (2014).csv')
df
```

```python
df.info()
```

```python
df['npH']=df.Ph.apply(lambda x: (100 if (8.5>=x>=7)  
                                 else(80 if  (8.6>=x>=8.5) or (6.9>=x>=6.8) 
                                      else(60 if (8.8>=x>=8.6) or (6.8>=x>=6.7) 
                                          else(40 if (9>=x>=8.8) or (6.7>=x>=6.5)
                                              else 0)))))
```

```python
#calculation of dissolved oxygen
df['ndo']=df.DO.apply(lambda x:(100 if (x>=6)  
                                 else(80 if  (6>=x>=5.1) 
                                      else(60 if (5>=x>=4.1)
                                          else(40 if (4>=x>=3) 
                                              else 0)))))
```

```python
#calc of B.O.D
df['nbod']=df.BOD.apply(lambda x:(100 if (3>=x>=0)  
                                 else(80 if  (6>=x>=3) 
                                      else(60 if (80>=x>=6)
                                          else(40 if (125>=x>=80) 
                                              else 0)))))
```

```python
df.rename(columns = {'TC(MPN/100mL)':'TC', 'FC(MPN/100mL)':'FC'}, inplace = True)
```

```python
#calculation of total coliform
df['ntc']=df.TC.apply(lambda x:(100 if (5>=x>=0)  
                                 else(80 if  (50>=x>=5) 
                                      else(60 if (500>=x>=50)
                                          else(40 if (10000>=x>=500) 
                                              else 0)))))
```

```python
#calculation of fecal coliform
df['nfc']=df.FC.apply(lambda x:(100 if (5>=x>=0)  
                                 else(80 if  (50>=x>=5) 
                                      else(60 if (500>=x>=50)
                                          else(40 if (10000>=x>=500) 
                                              else 0)))))
```

```python
df.head()
df.dtypes
```

```python
df['wph']=df.npH * 0.118
df['wdo']=df.ndo * 0.167
df['wbod']=df.nbod * 0.333
df['wtc']=df.ntc * 0.2
df['wfc']=df.nfc * 0.2
df['wqi']=df.wph+df.wdo+df.wbod+df.wtc+df.wfc 
df
```

```python
df.drop(df.iloc[:, 5:15], inplace = True, axis = 1)
```
